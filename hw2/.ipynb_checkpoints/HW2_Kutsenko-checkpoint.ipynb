## 1. Оцените дистанцию поездок (в метрах) на основе координат начальной и конечной станций (без учета поездок, завершившихся там же где и начались)

```python
columns_to_delete = [
    "tripduration",
    "starttime",
    "stoptime",
    "start station id",
    "start station name",
    "end station id",
    "end station name",
    "bikeid",
    "usertype",
    "birth year",
    "gender"
]

df_trips_uniq = df_trips.filter(
    df_trips["start station id"] != df_trips["end station id"]).\
    drop(*columns_to_delete).\
    withColumnRenamed("start station latitude", "StartLat").\
    withColumnRenamed("start station longitude", "StartLong").\
    withColumnRenamed("end station latitude", "EndLat").\
    withColumnRenamed("end station longitude", "EndLong")

print("Всего поездок: ", df_trips.count())
print("Уникальных поездок: ", df_trips_uniq.count())
```

```python
df_trips_uniq.show()
```

```python
!pip install haversine
```

```python
from haversine import haversine, Unit
```

```python
def haversine_m(lat1, long1, lat2, long2):
    return haversine((lat1, long1), (lat2, long2), unit=Unit.METERS)

haversine_udf = F.udf(haversine_m, returnType="double")
```

```python
df_trips_uniq_with_distances = df_trips_uniq.withColumn(
    "DistanceMeters",
    haversine_udf("StartLat", "StartLong", "EndLat", "EndLong")
)
```

```python
df_trips_uniq_with_distances.show()
```

## 2. Выведите максимальное, среднее значение, стандартное отклонение и медиан.

```python
column = "DistanceMeters"

stat_df = df_trips_uniq_with_distances.agg(
    F.max(column).alias("max"),
    F.avg(column).alias("mean"),
    F.stddev(column).alias("stddev")
)

median = df_trips_uniq_with_distances.approxQuantile(column, [0.5], 0.0)[0]

result = stat_df.withColumn("median", F.lit(median))
result.show()
```
